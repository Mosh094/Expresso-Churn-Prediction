# Fill up Missing Values

Categrorical Values

```python
# Fill up missing categorical values ('TOP_PACK' and 'REGION' ) with N/A
for i in ['TOP_PACK', 'REGION']:
    df_train[i] = df_train[i].fillna(value = 'N/A')
    df_test[i] = df_test[i].fillna(value = 'N/A')
```

Numerical Values

```python
# Fill up missing numerical values with the mean
for i in ['MONTANT', 'FREQUENCE_RECH', 'REVENUE','ARPU_SEGMENT', 'FREQUENCE', 'DATA_VOLUME', 'ON_NET', 'ORANGE', 'TIGO',
          'ZONE1', 'ZONE2', 'FREQ_TOP_PACK']:
    df_train[i] = df_train[i].fillna(value = 0)
    df_test[i] = df_test[i].fillna(value = 0)
```

Import LabelEncoder library 

The categorical data will be converted to numeric using LabelEncoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#For Train Dataset
le = LabelEncoder()
df_train["REGION"] = le.fit_transform(df_train["REGION"].astype(str))
df_train["TENURE"] = le.fit_transform(df_train["TENURE"].astype(str))
df_train["TOP_PACK"] = le.fit_transform(df_train["TOP_PACK"].astype(str))
df_train
```

```python
#For Test Dataset
le = LabelEncoder()
df_test["REGION"] = le.fit_transform(df_test["REGION"].astype(str))
df_test["TENURE"] = le.fit_transform(df_test["TENURE"].astype(str))
df_test["TOP_PACK"] = le.fit_transform(df_test["TOP_PACK"].astype(str))
df_test
```

```python
# Recollect MRG and user-id are to be dropped
X_train = df_train.drop(["user_id", "MRG", "CHURN"], axis =1)
y_train = df_train["CHURN"]

X_test = df_test.drop(["user_id", "MRG"], axis = 1)
```

# Build Model

Import Machine Learning Packages

```python
# ML Pkg
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

# Metrics
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

Split Training and Test Data

```python
x_train,x_validation,y_train,y_validation = train_test_split(X_train,y_train,test_size=0.30,random_state=7)
```

Using Nearest Neighnors

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
```

Model Accuracy

```python
knn.score(x_validation,y_validation)
```

```python
prediction = knn.predict(x_validation)
```

Evaluate Accuracy and Log_loss

```python

from sklearn.metrics import log_loss, accuracy_score
```

```python
score = accuracy_score(y_validation, prediction)
print(f"Log Loss is: {score}.")
```

Making Predictions on Test Dataset

```python
Test_prediction = knn.predict_proba(X_test)[:,1] #predicting probability.  showing the user_id column and the prediction alone
```

Save prediction file for submission

```python
pd.DataFrame({"user_id": df_test["user_id"], "CHURN": Test_prediction}).to_csv("Third_Submission.csv", index = False)
#Map the expected user_id column to user_id on the test file, and also the 'Test _prediction' to 'CHURN'
```
